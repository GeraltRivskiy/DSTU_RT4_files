```python
import matplotlib.pyplot as plt
import numpy as np
```

Самый простой график

```python
# создаем окно графика
fig, ax = plt.subplots()
# выводим самый простой график
ax.plot(np.linspace(0,15,25), np.random.randint(0,100,25))
```

Точечный график

```python
data = {'a': np.arange(50),
        'b': np.arange(50) + 10 * np.random.randn(50),
        'c': np.random.randint(0, 50, 50)*10}

fig, ax = plt.subplots(figsize=(6, 6), layout='constrained')
ax.scatter('a', 'b', s = 'c', data=data)
ax.set_xlabel('ось a')
ax.set_ylabel('ось b')
```

Несколько линий на одном графике

```python
x = np.linspace(0, 2, 100) #ось X

fig, ax = plt.subplots(figsize=(6, 4), layout='constrained' )
ax.plot(x, x, label='прямая')
ax.plot(x, x**2, label='квадрат')
ax.plot(x, x**3, label='куб')
ax.set_xlabel('ось X')
ax.set_ylabel('ось Y')
ax.set_title('Имя графика')
ax.legend
```

Несколько графиков в пределах одного окна

```python
def my_plotter(ax, data1, data2, param_dict):
    out = ax.plot(data1, data2, **param_dict)
    return out

data1, data2, data3, data4 = np.random.randn(4, 100)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

Стили линий

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(data1, 'o', label='один')
ax.plot(data2, 'd', label='два')
ax.plot(data3, 'v', label='три')
ax.plot(data4, 's', label='четыре')
ax.legend
```

Подписи данных

```python
mu, sigma = 115, 15 
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(6, 6), layout='constrained')

n, bins, patches =ax.hist(x, 50, density=1, alpha=0.75) 

ax.set_xlabel('ось X')
ax.set_ylabel('ось Y')
ax.set_title('Название графика\n в две строки')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True)
```

Указываем на что-то на графике

```python
fig, ax = plt.subplots(figsize=(6, 4))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2)

ax.annotate('локальный \n макcимум', xy=(2,1), xytext=(3, 1.5), arrowprops=dict(facecolor='blue'))

ax.set_ylim(-1.5, 2)
```
